```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r'https://raw.githubusercontent.com/AlexTheAnalyst/PandasYouTubeSeries/refs/heads/main/Ice%20Cream%20Ratings.csv')
```

```python
df=df.set_index("Date")
df
```

```python
print(plt.style.available)
plt.style.use('seaborn-v0_8-darkgrid')
```

```python
df.plot(kind='line',title="Ice Cream Ratings",xlabel="daily ratings" , ylabel="scores")
```

```python
df.plot.barh(stacked=True)
```

```python
df.plot.scatter(x="Texture Rating",y="Overall Rating",s=500,c='yellow')
```

```python
df.plot.hist(bins=150)
```
